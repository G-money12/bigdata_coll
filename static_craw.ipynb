```python
url="http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
data={'stnID':'159'}
r=requests.get(url,params=data)
r.encoding='utf-8'
webPage=r.text
soup=bs(webPage,'html.parser')
# print(soup.prettify())
```

```python
list1=[]
loc=soup.find_all("location")
for l in loc:
    city=l.city.string
    datas=l.find_all("data")
    # print(datas)
    for d in datas:
        tmef=d.tmef.string
        wf=d.wf.string
        tmn=d.tmn.string
        tmx=d.tmx.string
        # print(city,tmef,wf,tmn,tmx)
        list1.append((city,tmef,wf,tmn,tmx))   # 튜플로 저장
print(list1)
```

```python
import pandas as pd
df=pd.DataFrame(list1,columns=['city','tmef','wf','tmn','tmx'])
df.to_csv("data/weather.csv",index=False,encoding='utf-8-sig')
df
```

```python
books=soup.select('#yesSchList > li')
for book in books:
    author=book.select_one("span.authPub.info_auth > a").text
    print(author)
```

- title(제목)

```python
books=soup.select('#yesSchList > li')
for book in books:
    title=book.select_one("a.gd_name")
    if title is not None:
        title_txt=title.text
    print(title_txt)
```

- author(저자)

```python
books=soup.select('#yesSchList > li')
for book in books:
    author=book.select_one("span.authPub.info_auth > a")
    if author is not None:
        author_txt=author.text
    print(author_txt)
```

- publish(출판사)

```python
books=soup.select('#yesSchList > li')
for book in books:
    publish=book.select_one("span.authPub.info_pub > a")
    if publish is not None:
        publish_txt=publish.text        
    print(publish_txt)
```

- price(가격)

```python
books=soup.select('#yesSchList > li')
for book in books:
    price=book.select_one("strong > em")
    if price is not None:
        price_txt=price.text
    print(price_txt)
```

- grade(평점)

```python
books=soup.select('#yesSchList > li')
for book in books:
    grade=book.select_one("span.rating_grade > em")
    if grade is not None:
        grade_txt=grade.text
    print(grade_txt)
```
