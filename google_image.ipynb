```python
from selenium import webdriver      # <- 요놈 
from selenium.webdriver.common.by import By
import time
import urllib.request as req
```

google 이미지 검색 주소

```python
url="https://www.google.com/imghp"
query='손흥민'                        # 검색어 
driver=webdriver.Chrome()
driver.get(url)
time.sleep(3)
search_box=driver.find_element(By.NAME, 'q')
search_box.send_keys(query)
# search_box.send_keys(Keys.ENTER)
search_box.submit()
time.sleep(5)

PAUSE_TIME=2    # 스크롤 사이의 대기 시간
# 현재 페이지의 세로 스크롤의 위치
last_height=driver.execute_script("return window.scrollyY")     # 페이지의 초기 스크롤 높이 가져오기 
new_height=0
interval=2

# 현재 페이지 전체 스크롤 가능 높이
prev_height=driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(interval)

    curr_height=driver.execute_script("return document.body.scrollHeight")

    if curr_height==prev_height:
        break
    
    prev_height=curr_height

img_elements=driver.find_elements(By.CSS_SELECTOR,'div.H8Rx8c img.YQ4gaf')

save_path=f"C:/Users/it/Desktop/Python/bigdata_coll/data/google"        # \\ 두번적기 싫어서 r을 붙임, f을 쓰면 /을 써야합니다.
import os   # 검색어를 적어서 폴더안에 넣어둘것이다.
if not os.path.exists(save_path):  # save_path가 없다면 save_path 디렉토리를 만들어 달라.
    os.makedirs(save_path)

imgs=[]

for index, img in enumerate(img_elements):
    try:                                    # Error 생길때를 대비하여 try: , except: pass
        img_src=img.get_attribute('src')
        img_alt=img.get_attribute('alt')

        if img_src and "http" in img_src:
            imgs.append({'src': img_src, 'alt':img_alt}) 
            req.urlretrieve(img_src, f"{save_path}/{query}_{index}.png")
            print(f"{index}:다운로드 완료 - {img_alt}")
    except:
        pass
        #print(img_src)
        #print(img_alt)

driver.close()
```
