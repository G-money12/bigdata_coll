# Aladin 스크랩

```python
from bs4 import BeautifulSoup as bs
import requests
import re
import json
```

```python
base='https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page='     # 두번째, 세번째 등등 주소
#url='https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1'
books=[]
for i in range(3):
    url=base+str(i+1)
    res=requests.get(url)               # url 가져요기
    soup=bs(res.text,'html.parser')     # 뷰티플숲 형태로 파싱
    #print(soup)                        # 언제나 한번 확인하고

    book_list=soup.findAll('div',{'class':'ss_book_list'})
    for index, book in enumerate(book_list):                        
        if index % 2==0:
            print(f'index: {index*(i+1)}')
            book_ul=book.find('ul')     
            title=book_ul.find('a',{'class':'bo3'})
            if title !=None:
                title=title.text
            else:
                title="---"
               # class 가 bo3인 text(글자)인 것 찾기
            #price=book_ul.find('span',{'class':'ss_p'}).text
            print(f'title:{title}')
            #print(f'price:{price}')
            book_price=""
            book_author=""
            #print(book_li_list)  
                
            book_li_list=book_ul.findAll('li')
            price=book_li_list[3].find('span').text.replace(',','')
            print(f'price: {price}')
            author=book_li_list[2].find('a')
            if author !=None:
                author=author.text
            else:
                author=book_li_list[1].find('a').text            
            print(f'author: {author}')
            books.append((title, int(price), author))        # 튜플
            #print(book_ul)
            #break
        
```

data 프레임으로 저장됨

```python
import pandas as pd
df=pd.DataFrame(books,columns=['title','price','author'])
df.to_csv('books.csv',index=False)
#books
#len(books)
df
```
