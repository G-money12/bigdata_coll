- get 전체 가져오기

```python
url1="http://unico2013.dothome.co.kr/crawling/exam.html"
url2="http://unico2013.dothome.co.kr/crawling/post.php"
```

```python
import requests

r=requests.get(url1)
r.encoding='utf-8'
print(type(r))
print(r.text)
```

- head만 가져오기

```python
r=requests.head(url1)
r.encoding='utf-8'
print(type(r))
print(r.text)
```

- post 형태로 출력

```python
url2="http://unico2013.dothome.co.kr/crawling/post.php"
r=requests.post(url2,data={'name':'홍길동','age':12})
r.encoding='utf-8'
print(r.text)
```
